### Step 5. Create a histogram of the top 5 items bought

```python
top_five_items = chipo["item_name"].value_counts().head(5)
answer = chipo[chipo["item_name"].isin(top_five_items.index)]
plt.hist(answer["item_name"])
plt.title("Top 5 Items Bought")
plt.xlabel("Items")
plt.ylabel("Quantity Bought")
plt.xticks(rotation=45)

```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
plt.scatter(chipo["item_price"], chipo["quantity"] )
plt.title('Number of Items Ordered per Order Price')
plt.xlabel('Order Price')
plt.ylabel('Number of Items Ordered')
plt.xticks(rotation=90)


```
